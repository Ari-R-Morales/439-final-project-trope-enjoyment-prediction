```python
def patterns_to_raw_tags(patterns):
    return ", ".join(patterns)

for trope_id, patterns in blurb_match_rules.items():
    mask = tropes["trope_id"] == trope_id
    if mask.any():
        tropes.loc[mask, "raw_tags"] = patterns_to_raw_tags(patterns)
```
